```python
import re
import operator

text = ''
with open('tweets.txt', 'r', encoding = 'utf-8') as f:
    for line in f:
        text += line.strip() + ' '
        
lst_tags = re.findall('[#].\w*', text)
dct = {}
for i in lst_tags:
    if dct.get(i) == None:
        dct[i] = 1
    else:
        dct[i] += 1

lst_tags_sort = sorted(dct.items(), key = operator.itemgetter(1), reverse = True)

with open('tags.txt', 'w', encoding = 'utf-8') as f:
    for c in lst_tags_sort:
        print(f'тег: {c[0]}, упоминаний: {c[1]}', file = f)
```

Вычислить итоговую оценку студента по формуле (не забудьте преобразовать оценки из списка в числа!)
Напечатать в файл emails.txt первые три строки текста для рассылки в формате, указанном в условии. Используйте f-строки, чтобы подставить значения переменных в шаблон.
Проверить условие: если итоговая оценка меньше 4, то допечатать текст в рассылку о пересдаче в файле emails.txt. В ином случае допечатать пожелание хороших каникул.


```python
with open('students.csv', 'r', encoding = 'utf-8') as finput:
    with open('emails.txt', 'w', encoding = 'utf-8') as fout:
        for line in finput:
            fio = line.split(';')[0].strip()
            email = line.split(';')[1].strip()
            avg_oc = round(int(line.split(';')[2].strip()) * 0.4 + int(line.split(';')[3].strip()) * 0.6)
            print(f'To: {email}\nДобрый день, {fio}!\nВаша итоговая оценка за курс - {avg_oc}.', file = fout)
            if avg_oc < 4:
                print('Пересдача состоится 15 января в 10:00, место будет сообщено позднее.', file = fout)
            else:
                print('Хороших каникул!', file = fout)
        
```
