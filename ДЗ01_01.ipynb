## Долг
Алеша взял у Даши деньги в долг и обещал вернуть их с процентами.
Посчитайте, сколько денег получит Даша. 

ФОРМАТ ВВОДА 
На первой строке вводится сумма, которую дала Даша в долг. 

На второй строке вводится целое число — проценты, с которыми Алеша обещал вернуть долг. 

ФОРМАТ ВЫВОДА 
Целое число рублей — сумма денег, которую Даша должна получить от Алеши. Без копеек. 
Подсказка: 
чтобы посчитать проценты от суммы, разделите целое число процентов на 100 и потом умножьте сумму на получившуюся дробь. 
Например, чтобы посчитать 15 % от 500 рублей, нужно 15 разделить на 100 (получим 0.15), и умножить 500 на 0.15.


Для примера:

Ввод	Результат
1000
10
1100
500
15
575


```python
summa = int(input('Сколько Алеша взял в долг: '))
tax = int(input('Процент: '))



print(f'Даша должна получить {int(summa + summa * tax / 100)}')
```

```python
summa = int(input())
tax = int(input())

print(int(summa + summa * tax / 100))
```

## Алиса в течение года проверяет, какая температура в Сочи. 
Она хочет прилететь туда, 
когда температура там будет либо от +20 до +30 градусов Цельсия включительно, чтобы искупаться в море, 
либо когда температура будет от -5 до +5 градусов Цельсия включительно, чтобы ездить в горы и кататься на лыжах. 
Помогите Алисе определить, можно ли лететь в Сочи. 

ФОРМАТ ВВОДА 

Целое число — количество градусов по Цельсию в Сочи
ФОРМАТ ВЫВОДА 

Если Алиса может лететь в Сочи, выводится строка "Можно лететь!". 
Иначе выводится строка "Не время для отпуска".

Для примера:

Ввод	Результат
25      Можно лететь!


```python
degrees = int(input())

if -5 <= degrees <= 5:
    print('Можно лететь!')
elif 20 <= degrees <= 30:
    print('Можно лететь!')
else:
    print('Не время для отпуска')
```

## Банкомат
Банкомат выдает суммы, только кратные 100 рублям, потому что купюр мельче у него нет. Выведите на экран "ДА", если банкомат сможет выдать введенную сумму, и "НЕТ", если не сможет. 

ФОРМАТ ВВОДА
Вводится требуемая сумма для вывода средств, целое число без копеек. 
 ФОРМАТ ВЫВОДА

Ввод	Результат

НЕТ

## Помощник Айболита
Доктор Айболит решил собрать робота, который поможет ему делать стандартные назначения. Для начала робота нужно научить следующим действиям: 

Сначала спросить пациента, на что он жалуется 
Если пациент пожаловался на кашель, спросить, какой кашель — влажный или сухой? От влажного кашля робот назначает "Отхаркин-1", а от сухого "Отхаркин-2" 
Если пациент пожаловался на температуру, спросить, какая температура. Если она выше 38 градусов, робот назначает "Антижарин-1", иначе "Антижарин-2" 
Иначе робот отправляет пациента к доктору Айболиту 
Помогите Айболиту реализовать такого робота. 

ФОРМАТ ВВОДА 

Строка с первоначальной жалобой пациента. 
Если в качестве жалобы пациент ввел "кашель", с новой строки вводится характер кашля — строка "влажный" или сухой".
Если в качестве жалобы пациент ввел "температура", с новой строки вводится положительное вещественное число — температура пациента. 
ФОРМАТ ВЫВОДА 

Если робот знает, как обработать жалобу, выводится строка в формате "Назначаю вам <название лекарства>". 
В ином случае выводится строка "Пожалуйста, обратитесь к доктору Айболиту".


```python
zhaloba = input()
if zhaloba == 'кашель':
    zhaloba_type = input()
    if zhaloba_type == 'влажный':
        print('Назначаю вам Отхаркин-1')
    elif zhaloba_type == 'сухой':
        print('Назначаю вам Отхаркин-2')
    else:
        print('Пожалуйста, обратитесь к доктору Айболиту')
elif zhaloba == 'температура':
    zhaloba_temper = float(input())
    if zhaloba_temper > 38:
        print('Назначаю вам Антижарин-1')
    else:
        print('Назначаю вам Антижарин-2')
else:
    print('Пожалуйста, обратитесь к доктору Айболиту')
```

## Путевки на Олимпийские игры

```python
result_world = int(input()) 
result_euro = int(input())
if result_world < 0 or result_euro < 0:
    print("К сожалению, в этот раз не удалось завоевать путевки на Олимпийские игры")
else:
    if (result_world <= 20 and result_euro <= 16):
        print("Ура! Путевки на Олимпийские игры завоеваны!")
        if (result_world <= 2 or result_euro <= 2):
            print(3)
        elif (result_world <= 10 or result_euro <= 10):
            print(2)
        else: 
            print(1)

    elif (result_world <= 24 or result_euro <= 20):
        print("Нужно принять участие в дополнительных отборочных соревнованиях")

    else:
        print("К сожалению, в этот раз не удалось завоевать путевки на Олимпийские игры")

```

## Повышенная стипендия за отличную учебу равна 20000

```python
count = 0
summa = int(input())

while summa >= 0:
    if summa >= 20000:
        count += 1
    summa = int(input())

print(count)
```
