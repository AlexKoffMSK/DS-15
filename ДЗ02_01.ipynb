Функция filter_accounts, которая принимает в качестве параметра словарь accounts, ключами которого являются адреса аккаунтов, а значениями — число подписчиков. 
Также функция принимает в качестве параметров целые числа min_followers и max_followers, где min_followers меньше либо равен max_followers.

В этой задаче вы только определяете функцию: не нужно считывать ввод и вызывать ее, мы сделаем это автоматически при проверке.

ЧТО ИСПОЛЬЗУЕМ:

ЧТО ДЕЛАЕМ:
Создаем функцию filter_accounts, которая принимает три параметра — словарь accounts и два числа (min_followers и max_followers).
Внутри функции создаем пустой список.
Внутри функции запускаем цикл по словарю accounts. Для каждого ключа в аccounts делаем проверку: если количество подписчиков (значение ключа) находится в промежутке между минимальным количеством и максимальным (min_followers и max_followers), то добавляем аккаунт в список из п. 2.
Возвращаем отсортированный по алфавиту список из п. 2.
Для примера:

Ввод	Результат
{'@blogger13999': 2308, '@dragondragon': 1473, '@account': 1400, '@flowers': 1250, '@banana': 1500}, 1400, 1500
['@account', '@banana', '@dragondragon']




```python
# Вариант решения 1
def filter_accounts(accounts, min_followers, max_followers):
    s = []
    for account, count_follow in accounts.items():
        if min_followers <= count_follow <= max_followers:
            s += [account]
    s.sort()
    return s
```

```python
# Вариант решения 2
def filter_accounts(accounts, min_followers, max_followers):
    return sorted([account for account, count_follow in accounts.items() if min_followers <= count_follow <= max_followers])
```

```python
accounts = {'@blogger13999': 2308, '@dragondragon': 1473, '@account': 1400, '@flowers': 1250, '@banana': 1500}
min_followers = 1400
max_followers = 1500
```

```python
filter_accounts(accounts, min_followers, max_followers)
```
