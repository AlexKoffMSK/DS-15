Внутри функции создаем пустой список.
Внутри функции запускаем цикл по словарю accounts. Для каждого ключа в аccounts делаем проверку: если количество подписчиков (значение ключа) находится в промежутке между минимальным количеством и максимальным (min_followers и max_followers), то добавляем аккаунт в список из п. 2.
Возвращаем отсортированный по алфавиту список из п. 2.
Для примера:

Ввод	Результат
{'@blogger13999': 2308, '@dragondragon': 1473, '@account': 1400, '@flowers': 1250, '@banana': 1500}, 1400, 1500
['@account', '@banana', '@dragondragon']




```python
# Вариант решения 1
def filter_accounts(accounts, min_followers, max_followers):
    s = []
    for account, count_follow in accounts.items():
        if min_followers <= count_follow <= max_followers:
            s += [account]
    s.sort()
    return s
```

```python
# Вариант решения 2
def filter_accounts(accounts, min_followers, max_followers):
    return sorted([account for account, count_follow in accounts.items() if min_followers <= count_follow <= max_followers])
```

```python
accounts = {'@blogger13999': 2308, '@dragondragon': 1473, '@account': 1400, '@flowers': 1250, '@banana': 1500}
min_followers = 1400
max_followers = 1500
```

Внутри функции разделяем дату на три целых числа и сохраняем их в переменные — день, месяц, год.
Из модуля calendar вызываем фукнцию weekday() — передаем ей переменные с годом, месяцем и днем. Сохраняем результат (номер дня недели) в новую переменную.
Делаем проверку — если номер дня из п. 4 меньше 5, то возвращаем строку, что нужно работать. В остальных случаях выводим информацию о том, что день рождения в выходной день.
Для примера:

Ввод	Результат
'14-11-1995', 2020
Ура, выходной в День рождения!


```python
from calendar import weekday 
def is_birthday_weekend(date, year):
    day = int(date.split('-')[0])
    month = int(date.split('-')[1])
    if day == 29 and month == 2:
        day = 1
        month = 3
    if weekday(year, month, day) in (5, 6):
        return 'Ура, выходной в День рождения!'
    else:
        return 'В День рождения придется поработать.'
```

Для примера:

Ввод	Результат
['молоко; молочные продукты', 'помидоры; овощи', 'огурцы; овощи', 'творог; молочные продукты', 'сыр; молочные продукты']
{'молочные продукты': ['молоко', 'творог', 'сыр'], 'овощи': ['помидоры', 'огурцы']}




```python
def group_goods(lst_goods):
    dct = {}
    for string in lst_goods:
        string_lst = string.split('; ')
        if dct.get(string_lst[1]) == None:
                dct[string_lst[1]]=[string_lst[0]]
        else:
            dct[string_lst[1]].append(string_lst[0])
    return dct
```

## Метеостанция

Амплитуда температуры называется разница между максимальным и минимальным значением температуры за день. Напишите функцию, которая поможет сотрудникам метеостанции. 

ТРЕБУЕМАЯ ФУНКЦИЯ 

Функция weather_stat, которая принимает в качестве параметра список вещественных чисел observations. 
Функция должна вернуть минимальное, максимальное, среднее значения в списке (округлённое до 4 знака) и амплитуду.
В этой задаче вы только определяете функцию: не нужно считывать ввод и вызывать ее, мы сделаем это автоматически при проверке.

ЧТО ИСПОЛЬЗУЕМ:

ЧТО ДЕЛАЕМ:
Создаем функцию weather_stat, которая принимает в качестве параметра список вещественных чисел – observations.
Внутри функции находим и сохраняем в разные переменные:
Минимальное значение температуры.
Максимальное значение температуры.
Среднее значение температуры (складываем все элементы списка и делим на длину списка).
Амплитуду — разницу между максимумом и минимумом.
Верните все переменные в нужном порядке (используйте оператор return и после пробела перечислите все найденные значения через запятую.
Для примера:

Ввод	Результат
[-77.8874, -64.1202, 29.6946, 73.6522, 69.3968, -13.9534, -82.6001, 79.1074, -7.3194, 4.3518, 6.925, -69.4356, -7.127, -31.7698, 49.5321, -67.3312, -76.3036, -37.0821, 78.5122, 78.4804, 38.1691, -53.9157, 25.0968, -5.2388, -86.7001, -41.5536, 83.6603, 91.9444]
-86.7001, 91.9444, -0.4934, 178.6445



                                         

```python
def weather_stat(observation):
    summ = 0
    count = 0
    minn = 1000
    maxx = -1000
    for i in range(len(observation)):
        summ += observation[i]
        count += 1
        if observation[i] < minn:
            minn = observation[i]
        if observation[i] > maxx:
            maxx = observation[i]
    return(minn, maxx, round(summ/count, 4), maxx - minn)
```
